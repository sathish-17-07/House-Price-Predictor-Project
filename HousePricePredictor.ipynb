```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'],axis=1)
y = data['median_house_value']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.loc[:, train_data.columns!='ocean_proximity'].corr(), annot=True, cmap="YlGnBu")
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
dum = pd.get_dummies(train_data.ocean_proximity)
dum.astype(int)
```

```python
train_data = train_data.join(dum.astype(int)).drop(['ocean_proximity'],axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot = True, cmap = "YlGnBu")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude', y='longitude', data=train_data, hue="median_house_value", palette="coolwarm")
```

```python
from sklearn.linear_model import LinearRegression

x_train, y_train = train_data.drop(['median_house_value'],axis=1), train_data['median_house_value']

reg = LinearRegression()

reg.fit(x_train, y_train)
```

```python
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

#test_data = test_data.join(dum.astype(int)).drop(['ocean_proximity'],axis=1)
```

```python
dum = pd.get_dummies(test_data.ocean_proximity)
dum.astype(int)
```

```python
test_data = test_data.join(dum.astype(int)).drop(['ocean_proximity'],axis=1)
```

```python
test_data
```

```python
x_test, y_test = test_data.drop(['median_house_value'],axis=1), test_data['median_house_value']
```

```python
reg_test = reg.score(x_test, y_test)
```

```python
reg_train = reg.score(x_train, y_train)
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude', y='longitude', data=test_data, hue="median_house_value", palette="coolwarm")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest = RandomForestRegressor()

forest.fit(x_train, y_train)
```

```python
forest_test = forest.score(x_test, y_test)
```

```python
forest_train = forest.score(x_train, y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

decisionTree = DecisionTreeRegressor()

decisionTree.fit(x_train, y_train)
```

```python
dt_test = decisionTree.score(x_test, y_test)
```

```python
dt_train = decisionTree.score(x_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

knn = KNeighborsRegressor()

knn.fit(x_train, y_train)
```

```python
knn_test = knn.score(x_test, y_test)
```

```python
knn_train = knn.score(x_train, y_train)
```

```python
#Accuary of each alogirthm displaying as table
res_table = pd.DataFrame({
    'Model': ['Linear Regression','Random Forest Regression','Decision Tree Regression','KNN Regression'],
    'Train Score': [reg_train,forest_train,dt_train,knn_train],
    'Test Score' : [reg_test,forest_test,dt_test,knn_test]})
res_table.sort_values(by='Train Score', ascending=False)
```

```python
#Evaluating Random Forest model using MSE, RMSE and MAE

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

y_pred = forest.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mean_actual = y_test.mean()
mae_percentage = (mae / mean_actual) * 100

mse = mean_squared_error(y_test, y_pred)
var_actual = y_test.var()
mse_percentage = (mse / var_actual) * 100

rmse = np.sqrt(mse)
range_actual = np.ptp(y_test)
rmse_percentage = (rmse / range_actual) * 100


#print('Mean Absolute Error',mae_percentage)
#print('Mean Squared Error',mse_percentage)
#print('RMSE',rmse_percentage)

res_table = pd.DataFrame({
    'Error Model': ['RMSE','Mean Squared Error','Mean Absolute Error'],
    'Error percentage': [rmse_percentage,mse_percentage,mae_percentage]
    })
res_table.sort_values(by='Error percentage', ascending=True)
```
